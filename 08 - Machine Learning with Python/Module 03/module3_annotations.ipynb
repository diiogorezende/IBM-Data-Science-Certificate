# KNN (K-Vizinhos Proximos)
Vamos supor que possuimos 4 classes, e queremos realizar a classificacao dos nossos clientes. Por exemplo, tendo informacoes demograficas, de salario, situacao empregaticia, etc... Podemos classificar e recomendar qual o servico mais adequado a se ofertar para o cliente.

O algoritmo KNN trabalha verificando a distancia do dados desconhecidos com relacao aos conhecidos. Mas quantos pontos de dados proximos sao necessarios para realizar uma classificacao?

Para isso, definimos a quantidade de "vizinhos proximos" que queremos verificar para realizar a classificacao. Suponha que definimos 5 vizinhos proximos, sendo assim, o algoritmo ira verificar qual eh a classe majoritaria desses 5 e, apos, ira classificar o novo dado de acordo com essa informacao.

**Como o algoritmo funciona**
- Definimos um K valor.
- Calcula-se a distancia do novo dado com relacao aos dados conhecidos.
- Selecionamos as K-observacoes nos dados de treino que sao proximas ao dado desconhecido.
- Realizamos a predicao do dado desconhecido usando o valor mais frequente dos K-vizinhos proximos.

# Metricas de Avaliacao para Problemas de Classificacao

## Indice Jaccard
y = rotulos conhecidos

y_hat = Rotulos previstos

Calcula a acuracia atraves do: Tamanho da interseccao de y e y_hat, dividido pelo modulo de y + y_hat - interseccao de y e y_hat.

Quanto mais proximo de 1 for o indice, mais acertos o modelo esta tendo.

## F1-Score
Primeiro, precisamos saber os valores do Precision e do Recall para calcular o F1-Score.

- Precision = TP / (TP + FP)
- Recall = TP / (TP + FN)

O F1-Score eh a media harmonica da precisao e do recall.
- F1-Score = 2 * (precision * recall) / (precision + recall)

## Log-Loss
Mede a performance de um classificador onde o valor previsto eh uma probabilidade entre 0 e 1.

# Introducao a Arvore de Decisao
A intuicao basica por tras de uma arvore de decisao eh mapear todos os caminhos decisorios possiveis em um formato de arvore.

As arvores de decisao sao construidas por Nodos, que contem uma categoria de dado. O nodo faz referencia a um teste, por exemplo, verificar o sexo (Masculino ou Feminino) de uma pessoa. Ja a Branch diz respeito ao resultado deste teste. E cada Leaf determina uma classificacao.

**Etapas de aprendizagem:**
- Realiza a escolha dos atributos
- Calcula a significancia dos atributos
- Divide os dados baseado no valor dos melhores atributos
- Volta ao passo inicial.

## Construcao de uma Arvore de Decisao
As arvores de decisao sao construidas usando Particao Recursiva para classificar os dados. Entao, seleciona os melhores atributos que irao realizar a divisao dos dados. 

Quando uma caracteristica fornece uma grande capacidade de separacao dos dados, dizemos que ela eh mais preditiva, menos impura, e possui menor entropia.

**Entropia:**
Eh a quantidade de informacao desordenada. Ela mede a aleatoriedade ou incerteza dos dados para uma determinada variavel, e tambem verifica o quanto a variavel explica os dados.

Quanto menor a entropia, menos uniforme eh a distribuicao, e mais puro eh o Node.

Apos calcular a Entropia, ainda precisamos calcular o **Ganho de Informacao**.

**Ganho de Informacao:**
Eh a informacao que aumenta o nivel de certeza depois da separacao dos dados. Conforme a entropia diminui, o Ganho de Informacao aumenta, e vice-versa.

**OBS:**
Ou seja, precisamos calcular a entropia para cada possibilidade da variavel em questao, e depois realizar o calculo do ganho de informacao.
